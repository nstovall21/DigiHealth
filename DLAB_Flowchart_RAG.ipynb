```python
import getpass
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS
import google.generativeai as genai
from langchain_google_genai import ChatGoogleGenerativeAI

#load file to be used
txt_file_path = '/content/drive/MyDrive/NewFlowCharts copy.txt'
loader = TextLoader(file_path=txt_file_path, encoding="utf-8")
data = loader.load()

#split text
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
data = text_splitter.split_documents(data)

#translate data into vectors
embeddings = OpenAIEmbeddings()

#FAISS to store and retreive vectors
vectorstore = FAISS.from_documents(data, embedding=embeddings)

# Create an instance of the LLM, using the 'gemini-pro' model with a specified creativity level
llm = ChatGoogleGenerativeAI(model='gemini-pro', temperature=0.7)

#allows conversations to be remembered and used for continued use
memory = ConversationBufferMemory(memory_key='chat_history', return_messages=True)

#ties knowledge to chatbot
conversation_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    chain_type="stuff",
    retriever=vectorstore.as_retriever(),
    memory=memory
)
```
